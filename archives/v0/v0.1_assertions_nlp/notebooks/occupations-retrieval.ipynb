```python
from os import getenv
import pandas as pd, numpy as np
import sqlalchemy
import geovpylib.database as db
import knex
import gmpykit as kit
from gmpykit.ipython import infos

db.connect_yellow('switzerland_and_beyond')
eta = kit.Eta()
```

```python
assertions = db.query('select * from hls.person_assertion;')
jobs = assertions[assertions['topic'] == 'job']

infos(jobs)
```

```python
occupations = set()
eta.begin(len(jobs), "Extract occupations")
for text in jobs['assertion']:
    # text = jobs.iloc[i]['assertion']
    # response = kit.ask_llm(f'Can you give me the job name mentioned in the following text? If it is not clear or unknown, or uncertain, just answer "No clear occupation". The text is: "{text}"')
    response = kit.ask_llm(f'What is the job mentioned in the following text? Answer with only the occupation name. Answer in english. Text: "{text}"')
    occupations.add(response)
    eta.print(response)
    eta.iter()
eta.end()
```

```python
for occupation in occupations:
    print(f"'{occupation}',")
```
